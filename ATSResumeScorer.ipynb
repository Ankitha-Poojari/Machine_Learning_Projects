```python
import pdfplumber
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text() + ' '
    return text.lower()

def preprocess_text(text):
    """Tokenize, remove punctuation and stopwords."""
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return set(filtered_tokens)

def calculate_match_score(resume_tokens, job_tokens):
    """Calculate ATS match score based on keyword overlap."""
    matched_keywords = resume_tokens.intersection(job_tokens)
    score = (len(matched_keywords) / len(job_tokens)) * 100 if job_tokens else 0
    return round(score, 2), matched_keywords

def main():
    job_description = input("Enter Job Description:\n").lower()
    resume_path = input("Enter Resume PDF File Path:\n")

    job_tokens = preprocess_text(job_description)
    resume_text = extract_text_from_pdf(resume_path)
    resume_tokens = preprocess_text(resume_text)

    score, matched_keywords = calculate_match_score(resume_tokens, job_tokens)

    print(f"\nATS Resume Match Score: {score}%")
    print(f"Matched Keywords: {', '.join(matched_keywords)}")

if __name__ == "__main__":
    main()
```
